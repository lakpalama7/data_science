# Map and Filter

```python
# map (function, iterable)
num=[1,2,3,4,5]
```

```python
def sqr(x):
    return x**2
    
```

```python
result = map(sqr, num)
print(list(result))
```

```python
result = map(lambda x: x**2, num)
print(list(result))
```

# Filter

```python
# filter(function, iterable)
numbers = list(range(2,99,3))
print(numbers, end='')
```

```python
# get only the even numbers
def check(x):
    if x %2 == 0:
        return x
```

```python
result = list(filter(check,numbers))
result
```

```python
# get the square of even numbers
def sqr(x):
    if x % 2 == 0:
        return x
```

```python

result = list(filter(sqr, numbers))
result
```

```python
sqr_even = list(map(lambda x: x**2, result))
print(sqr_even,end='')
```

```python
# map and filter to get square of even number
result=list(map(lambda x:x**2, filter(lambda x:x%2==0, numbers)))
print(result,end='')
```

# List comprehension

```python
given = list(range(1,11))
given
```

```python
# get square
sqr=[]
for x in given:
    sqr.append(x**2)
    
```

```python
sqr
```

```python
# use list comprehension
output = [x**2 for x in given]
output
```

```python
# list comprehension with condition
# from the list of numbers 1 to 100, create a list div by 6
div6 = [x for x in range(1,100) if x % 6 == 0]
print(div6, end='')
```

```python
# in a list of number 1 to 200 , create list of numbers divisible by 5 and 10
result = [x for x in range(1,200) if x % 5 == 0 and x %10 == 0]
print(result, end='')
```

```python
# list of numbers 1 to 200, create a list of numbers divisible by 5 and 15
result = [x for x in range(1,200) if x % 5 == 0 and x % 15 == 0]
print(result, end='')
```

# List slicing

```python
# list slicing is the process of getting a protion (slice) of list such that it doesnot looses the property of original list
```

```python
nums = list(range(2,31))
print(nums)
```

```python
nums[0]
```

```python
type(nums[0:2])
```

```python
# get index from 2 to 10 
nums[2:10]
```

```python
# get last 3 value
nums[-3:]
```

```python
# get portion from 2 to 10 in step 3
nums[2:10:3]
```

```python
# get all in step 2
nums[::2]
```

```python
# reverse
nums[-1:-4:-1]
```

```python
# reverse for 10 values in step 2
nums[-1:-10:-2]
```

```python
# reverse all 
print(nums[::-1])
```

```python
# reverse in step 3
print(nums[::-3])
```
