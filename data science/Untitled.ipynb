# map and Filter

```python
# map(functions, iterable) : each element expression is performed
```

```python
num=list(range(1,100,3))
print(num)
```

```python
# find the squre of each numbers
def val(x):
    return x**2
    
  
```

```python
result = map(val, num)
print(list(result))
```

```python
# using lamda fucntion
result = map(lambda x: x**2, num)
print(list(result))
```

```python
# Filter : only perform experession for filter data
# filter(fun,iterable)
```

```python
num = list(range(1,200,3))
print(num)
```

```python
# filter the numbers divsible by 5 and 7
def val(x):
    r5 = x % 5
    r7 = x % 7
    if r5 == 0 or r7 == 0:
        return x
    
```

```python
result = list(filter(val,num))
print(result)
```

```python
# find the even numbers from the filter result
final_result = list(map(lambda x: x**2, result))
print(final_result)
```

# LIst comprehension

```python
# list(expression, loop)
# list(expression, loop, condition)
```

```python
# list(expression, loop)
# find the squre of each number
num = list(range(1,10))
print(num)
```

```python
l=[x**2 for x in num]
print(l)
```

```python
s = list('kathmandu')
s
```

```python
r = [x.capitalize() for x in s]
r
```

```python
# list(expression, loop, condition)
num = list(range(1,99,3))
print(num)
```

```python
# find the square of even number only
r = [x**2 for x in num if x%2==0]
print(r)
```

```python
# find the square of number if number is divisible by 2 and 5
r = [x for x in num if x % 2 == 0 and x % 5 == 0]
print(r)
```
