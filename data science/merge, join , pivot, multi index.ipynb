```python
data['State'].unique()
```

```python
data.groupby('State')['Sales'].sum()
```

```python
# pivot table -- pivot_table()
pd.pivot_table(data=data, values='Sales', index='State', aggfunc='sum')
```

```python
data.Category.unique()
```

```python
pd.pivot_table(
    data=data,
    values='Sales',
    index='State',
    columns='Category',
    aggfunc=['min','max']
)
```

```python
data.head(2)
```

```python
pd.pivot_table(
    data=data,
    values='Sales',
    index='State',
    columns=['Category','Sub-Category'],
    aggfunc='sum'
)
```

```python
# create pivot table to show sales summary of each customers
pd.pivot_table(
    data=data,
    values='Sales',
    index=['Region','State','Sub-Category'],
    columns='Category',
    aggfunc='sum'
)
```
