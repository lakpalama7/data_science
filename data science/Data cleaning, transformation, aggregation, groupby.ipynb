```python
# >='very powerfull', >=1000 'powerfull' , >=500 'less powerfull'
# create funciton
def classify(value):
    if value >=2000:
        return 'Very Powerfull'
    elif value >=1000:
        return 'Powerfull'
    else:
        return 'less Powerfull'
```
