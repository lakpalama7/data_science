```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load pre build dataset
data = sns.load_dataset('tips')
data.head()
```

```python
# bar plot
sns.barplot(x='day', y='tip', data=data)
```

```python
# countplot
sns.countplot(x='time', data=data)
```

```python
# subplot
fig , axes = plt.subplots(1,2, figsize=(10,6))
sns.barplot(x='day', y='tip', data=data, ci=None, ax=axes[0], label='Barplot')
sns.countplot(x='time', data=data, ax=axes[1], label='countplot')
axes[0].set_title("Barplot Avg tip per day")
axes[1].set_title("Countplot of Time")
plt.legend()

```

```python
fig, axes = plt.subplots(2,2, figsize=(12,8))
sns.lineplot(x='day',y='tip', data=data, ax=axes[0,0], errorbar=None, label='day - tip', )
sns.countplot(x='sex', data=data, ax=axes[0,1], label='sex')
sns.barplot(x='day', y='total_bill', data=data, ax=axes[1,0], errorbar=None,palette='cool' )
sns.barplot(x='smoker', y='size', data=data, ax=axes[1,1],errorbar=None, label='smoker - size', palette='bright')
axes[0,0].set_title("Line chart of Avg tips daywise")
axes[0,1].set_title("Count chart of sex")
axes[1,0].set_title("Bar chart day and total bill")
axes[1,1].set_title("Bar chart of smoker and size")
plt.legend()
plt.tight_layout()
plt.savefig('demo1.png')
```

```python
# Before analysis
# find Outliers and remove
# using IQR, z-score , mean & std
```

```python
# Using IQR method - to find outliers and remove
```

```python
q1 = data['Revenue'].quantile(0.25)
q1
```

```python
q3 = data['Revenue'].quantile(0.75)
q3
```

```python
iqr = q3 - q1
iqr
```

```python
lower_bound = q1 - 1.5 * iqr
lower_bound
```

```python
upper_bound = q3 + 1.5 * iqr
upper_bound
```

```python
sns.kdeplot(x='Revenue', data=data)
plt.axvline(upper_bound, color='red', linestyle='-')
plt.axvline(lower_bound, color='green', linestyle='--')
```

```python
df = data[data['Revenue']<=upper_bound]
```

```python
#ploting only data below upper_bound
fig, axes = plt.subplots(1,2, figsize=(10,6))
sns.kdeplot(x='Revenue', data=data, ax=axes[0])
axes[0].set_title("KDE of Revenue before Removing Outliers")
sns.kdeplot(x='Revenue', data=df, ax=axes[1])
axes[1].set_title("KDE of Reveneue after Removing Outliers")
plt.tight_layout()
```
