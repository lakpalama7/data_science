# SET comprehension

```python
# creating set using loop ( comprehension)
```

```python
# creating a set of cube of numbers from 1 to 100 in range 3
print(list(range(1,100,3)))
```

```python
cube_set = {x**3 for x in range(1,100,3)}
print(cube_set)
```

```python
# create set with cube number for value divisible by 2
cube_set = {x**3 for x in range(1,100,3) if x % 2 == 0}
print(cube_set)
```
