```python
# >=500 'Les power', >=1500 ,'Powerfull', >=300, 'High power'
#create a fucntions
def classify(value):
    if value >=3000:
        return 'High Powerfull'
    elif value >=1500:
        return 'Powerfull'
    else:
        return 'Low Powerfull'
        
```

```python
data['Engine_Type']=data.Engine.apply(classify)
```
