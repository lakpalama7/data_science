```python
def age_group_creation(row):
    if row['age'] < 30:
        return "20 - 30"
    elif row['age']< 40:
        return "30 - 40"
    elif row['age'] < 50:
        return "40 -50"
    elif row['age'] < 60:
        return "50 -60"
    else:
        return "60-70"
```

```python
data.columns
```

```python
data['Age Range'] = data.apply(age_group_creation, axis=1)
```

```python
X = data[['age','annual Salary','net worth']]
y = data['car purchase amount']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
# split data into train and test
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
len(X_train)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_train
```

```python
import joblib
```

```python
joblib.dump(scaler, 'Scaler.pkl')
```

```python
X_test = scaler.fit_transform(X_test)
X_test
```

```python
len(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
def modelresults(prediction):
    print("Mean absolute error on model is {}".format(mean_absolute_error(y_test,prediction)))
    print("Mean square error on model is {}".format(np.sqrt(mean_squared_error(y_test,prediction))))
    print("R2 score on model is {}".format(r2_score(y_test,prediction)))
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
prediction_lr_model = lr.predict(X_test)
prediction_lr_model
```

```python
modelresults(prediction_lr_model)
```
